# One Hot Encoding

Implementación de One Hot Encoding en un data set relacionado a Recursos Humanos, en las variables de "Sexo" y "Ciudadanía".

```python
import pandas as pd
import numpy as np
```

```python
dfhr = pd.read_csv(filepath_or_buffer = "/Users/lalemus/Desktop/Escuela de Bayes/Business Analytics/Recursos Humanos/HRDataset_v13.csv")
```

```python
dfhr.head()
```

```python
dfhr1 = dfhr[["Employee_Name","PayRate","Position","State","Sex","DOB","MaritalDesc",
             "CitizenDesc","Department","PerformanceScore","EngagementSurvey","EmpSatisfaction"]]
```

```python
dfhr1.shape
```

```python
dfhr1.describe()
```

```python
dfhr1.head()
```

```python
dummy_sex = pd.get_dummies(dfhr1["Sex"], prefix = "sex")
```

```python
dummy_sex.head()
```

```python
dfhr1 = dfhr1.drop(["Sex"], axis = 1)
```

```python
dfhr1.shape
```

```python
dfhr2 = pd.concat([dfhr1,dummy_sex], axis = 1)
```

```python
dfhr2.shape
```

```python
dfhr2.head()
```

```python
def createdummies(df,varname):
    dummy = pd.get_dummies(df[varname], prefix = varname)
    df = df.drop(varname, axis = 1)
    df = pd.concat([df, dummy], axis = 1)
    return df
```

```python
createdummies(dfhr2, "CitizenDesc")
```

### Conclusión

En el último resultado podemos observar la implementación completa del ONE HOT ENCODING en las variables de "Sexo" y "Ciudadanía"
